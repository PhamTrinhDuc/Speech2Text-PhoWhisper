### Audio5 - 17s
```vào thứ bảy hàng tuần học sinh sẽ cùng nhau tham gia sinh hoạt tại câu lạc bộ thể thao với nhiều hoạt động tập luyện giao lưu thi đấu dưới sự dẫn dắt của các huấn luyện viên giàu kinh nghiệm đây là sân chơi để học sinh phát huy tài năng nâng cao tinh thần đồng đội rèn luyện kỹ năng kết nối chia sẻ và trưởng thành.```

````python


transcriber = pipeline("automatic-speech-recognition", model = "vinai/PhoWhisper-medium", device = "cuda")
text = transcriber("/content/drive/MyDrive/Colab Notebooks/Speech2Text/audio5.mp3")["text"]
print(text)
````
